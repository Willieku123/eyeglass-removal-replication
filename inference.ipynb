# Inference

Before inference, you can either collect your trained `latest_net_deglass.pth`, `latest_net_deshadow.pth`, `latest_net_G_DA.pth`, `latest_net_G_GLASS_MASK.pth` and `latest_net_G_SHADOW_MASK.pth` from checkpoint folders, or download our pre-trained weights and put them under `./checkpoints/fully_trained` .

`--input_root`:  Infer images in this directory. Remeber to resize images to 256x256 in advance!

`--pth_root`:  Use trained weights in this directory. 

`--output_root`:  Save images to this directory.

`--verbose`:  Also save glass/shadow masks and de-shadowed images.

`--batch_size`:  batch size
